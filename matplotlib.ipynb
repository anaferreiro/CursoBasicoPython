# Matplotlib

Librería de representación gráfica

```python
#cálculo del espacio a partir de la velocidad y la aceleración
#s(t)=v0*t+0.5* a*t**2
import matplotlib.pyplot as plt
import numpy as np
v0=0.2; a=2.

n=21 #num de valores de t donde vamos a representar
t0=0.; tf=2.
t=np.linspace(t0,tf,n)
s=v0*t + 0.5*a*t**2
fig=plt.figure()
plt.plot(t,s,'r-')
plt.savefig('migrafica.png')
plt.show() #no hace falta si estoy en jupyter
```

```python
#cálculo del espacio a partir de la velocidad y la aceleración
#s(t)=v0*t+0.5* a*t**2
import matplotlib.pyplot as plt
import numpy as np
v0=0.2; a=2.

n=21 #num de valores de t donde vamos a representar
t0=0.; tf=2.
t=np.linspace(t0,tf,n)
s=v0*t + 0.5*a*t**2
fig=plt.figure()
plt.plot(t,s,'r-', label='v0=0.2')
v0=0.8
s=v0*t + 0.5*a*t**2
plt.plot(t,s,'b.', label='v0=0.8')
plt.legend()
plt.savefig('migrafica.png')
plt.show() #no hace falta si estoy en jupyter
```

```python
# con funciones

def movement(t, v0, a):
    s=v0*t + 0.5*a*t**2
    v=v0 + a*t
    return s,v
    #return [s,v]

v0=0.2; a=2.
n=21 #num de valores de t donde vamos a representar
t0=0.; tf=2.
t=np.linspace(t0,tf,n)
val=movement(t,v0,a)
s=val[0]; v=val[1]

fig=plt.figure()
plt.plot(t,s,'r-', label='s')
plt.plot(t,v,'b.', label='v')
plt.title("velocidad= "+str(v0))
plt.legend()


v0=0.8; a=2.
fig=plt.figure()
plt.plot(t,s,'r-', label='s')
plt.plot(t,v,'b.', label='v')
plt.title("velocidad= "+str(v0))
plt.legend()

plt.show() 
```

```python
v0=0.2; a=2.
n=21 #num de valores de t donde vamos a representar
t0=0.; tf=2.
t=np.linspace(t0,tf,n)
val={}
listv0=[0.2,0.8]
for v0 in listv0:
    val[v0]=movement(t,v0,a)
    
lc={}
lc[listv0[0]]=['b-','o-']
lc[listv0[1]]=['r.','g:']

for key in val: 
    plt.subplot(211)
    plt.plot(t, val[key][0], lc[key][0])
    plt.title('Espacio')
    plt.ylabel('s')

    plt.subplot(2, 1, 2)
    plt.plot(t, val[key][1], lc[key][1])
    plt.xlabel('time (s)')
    plt.ylabel('v')

plt.show()
```
