# Lectura y Escritura de ficheros Excel con Pandas

```python
import pandas as pd
#fichero .csv cuyos datos están separados por comas. La primera fila indicaa la información que hay por cada columna
file='/content/drive/MyDrive/Ficheros_leer/notas.xlsx'
outfile = open(file, 'r')
#leemos el fichero complete
#dataset = pd.read_excel(file) #podemos indicar solo una hora del fichero con name_sheet
dataset = pd.read_excel(file,index_col='NIF')
print(dataset)
print(type(dataset))

print("***Indeces\n", dataset.index)
print("***Cabeceras\n",dataset.head())
#print("NIF\n",dataset['NIF']) #si NIF va como index_col, entonces NIF no se puede usar como clave
print("Nombre: \n", dataset['Nombre'])
print("P1:\n", dataset['P1'])




```

```python
for k in dataset:
  print(dataset[k])  #imprime las columnas



#Muestra las cinco primeras filas
print("Cinco primeras filas")
print(dataset.head())
#Muestra las cinco últimas filas
print("cinco últimas filas")
print(dataset.tail())
```

```python
#vamos a construir un diccionario, cuyas claves sean los NIFs, y asociado a cada clave las notas y las medias
alumnos={}
for nif in dataset.index:
  alumnos[nif]=[]

for d in dataset:
  print("----")
  print(d.index)
  print(d)
  print(dataset[d])
  for nif in dataset[d].index:
    #alumnos[nif].append({d:dataset[d][0]})
    if d !='Nombre':
      alumnos[nif].append(dataset[d][0])

print(alumnos)
import numpy as np
#vamos a incluir la media al final
for key in alumnos:
  media=np.mean(alumnos[key])
  alumnos[key].append(media)

for k in alumnos:
  print("k= ", alumnos[k])
```

```python
#guardamos nuestro diccionario en un fichero excel

file2='/content/drive/MyDrive/Ficheros_leer/notas_mod.xlsx'
outfile2 = open(file2, 'w')
df = pd.DataFrame(alumnos).T  # transpose to look just like the sheet above
df.columns=["NIF", "P1", "P2", "P3", "Media"]
df.to_excel(file2)
outfile2.close()
```
