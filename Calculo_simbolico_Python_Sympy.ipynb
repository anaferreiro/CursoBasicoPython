
Considerando $a=0.2$ y $b=10^4$, y el tiempo se mide en horas.

$$
y(t)= \frac{b}{1+(b-100)*10^{-2}*e^{-\frac{at^3}{3}}}
$$

Considerando $a=0.2$ y $b=150$, y el tiempo se mide en horas.

```python
import sympy as spy

t=spy.Symbol('t')
b=150; a=0.2
y=b/(1+(b-100)*1.e-2*spy.exp(-a*t**3/2))
print(y)

print("y(2)=", y.subs(t,2)," , ", y.subs(t,2).evalf() )
spy.plot(y)
```

```python
## Derivada
dy=spy.diff(y,t)
print(dy)
spy.plot(dy)
d2y=spy.diff(dy,t)
print("derivada segunda=\n ", d2y)
#tambien vale:
d2y=spy.diff(y,t,2)
print("Derivada segunda n",d2y)
```

```python
#Integrales
int_y=spy.integrate(y,t, manual=True)
print("Integral= ", int_y)  #no es capaz de hacerla
```

```python
#Integrales
x=spy.Symbol('x')
f=x**2*spy.exp(x)*spy.cos(x)
int_f=spy.integrate(f,x)
print("Integral indefinida=\n",int_f)
int2_f=spy.integrate(f,(x,-1,1))
print("Integral definida=\n",int2_f)
```
