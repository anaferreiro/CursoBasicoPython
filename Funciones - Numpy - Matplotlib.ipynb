Ecuación que describe el movimiento cuando consideramos $a(t)=a_0$ para $t\in [0,t_1]$ y $a(t)=0$ para $t>t_1$

$$
s(t)=\left\{\begin{array}{lcl}
s_0+v_0 t + \frac{1}{2}a_0t^2 & si & t\in[0,t_1]\\
s_0+v_0 t + \frac{1}{2}a_0t^2+ a_0t_1(t-t_1)& si & t>t_1\\
\end{array}
\right.
$$

```python
import numpy as np
import matplotlib.pyplot as plt

#OJO!! Con este modo de implementación no puedo trabajar sobre arrays de numpy
#La implemetación es totalmente válida, simplemente tengo que tener en cuenta que ahora 
#ya no puedo hacer actuar la función sobre todos los elemenots de un array

def s(t,s0,v0,a0,t1):
    if t<=t1: return s0+v0*t + 0.5 * a0*t**2
    elif t>t1: return s0+v0*t+ 0.5*t1**2+a0*t1*(t-t1)
    else: return None
```

```python
tf=10; t1=3.5; v0=0.2; a0=2.4;s0=2.
t=np.linspace(0,tf,120)
vt=np.zeros_like(t)

for k in range(0,t.size):
    vt[k]=s(t[k],s0,v0,a0,t1)

fig=plt.figure()
plt.plot(t,vt,'r-.')
plt.title("espacio")
plt.xlabel('t')
plt.ylabel('y')
plt.show()
```
